```python
import pandas as pd

df = pd.read_csv('DataAllMethods.csv', index_col=0)

#dfmgType = df[(df['microGridType']=='mg1') & (df['Size']==1000)]
dfmgType = df[(df['microGridType']=='mg1') & (df['Size']==50) & ((df['OptMethod'] != 'Scipy_L-BFGS-B') & (df['OptMethod'] != 'Scipy_SLSQP')) ]
for name in df.columns[4:]:
    boxplot = dfmgType.boxplot(column=name, by='OptMethod')
    if name=='ObjFunc':
        boxplot.set_ylim([0.27, 0.33])
```

```python
dfmgType = df[(df['microGridType']=='mg2') & (df['Size']==1000) & ((df['OptMethod'] != 'Scipy_L-BFGS-B') & (df['OptMethod'] != 'Scipy_SLSQP')) ]
for name in df.columns[4:]:
    dfmgType.boxplot(column=name, by='OptMethod')
```

```python
for method in df['OptMethod'].unique():
    dfmgType = df[(df['microGridType']=='mg2') & (df['OptMethod']==method) ]
    axes = dfmgType.boxplot(column='ObjFunc', by='Size')
    axes.set_title(method)
    axes.set_ylabel('ObjFunc')
```
